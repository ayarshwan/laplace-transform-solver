```python
import sympy as sp

# Define symbols globally
t, s = sp.symbols('t s')

print("Hello! If you want to find the Laplace transform of a function, please make sure that you write the function correctly:")
print("- Power: t**2 (for t squared)")
print("- Exponent: exp(-2*t)")
print("- Sine: sin(t)")
print("- Cosine: cos(t)")
print("- Logarithm (Natural Log): ln(t)")
print("- Combine operations, e.g.: exp(-2*t)*sin(3*t)")
print("- For integration, use: integrate(t**2, t) or integrate(t**2, (t, 0, 1))")
print("- For Laplace transform of shifted functions (Unit Step), use: u(t-3)*t**2 (where 'u' represents the unit step function)")

# Class for Laplace Transform
class LaplaceTransform:
    @staticmethod
    def compute(function_str):
        try:
            f_t = sp.sympify(function_str)
            laplace_result = sp.laplace_transform(f_t, t, s, noconds=True)
            return f"\nInput Function: {f_t}\nLaplace Transform: {laplace_result}"
        except Exception as e:
            return f"Error: {e}"

# Class for Inverse Laplace Transform
class InverseLaplaceTransform:
    @staticmethod
    def compute(function_str):
        try:
            f_s = sp.sympify(function_str)


            if 'ln' in function_str:
                f_s_decomposed = f_s
            else:
                try:
                    f_s_decomposed = sp.apart(f_s)
                except Exception:
                    f_s_decomposed = f_s

            inverse_laplace_result = sp.inverse_laplace_transform(f_s_decomposed, s, t)


            inverse_laplace_result = sp.simplify(inverse_laplace_result)
            inverse_laplace_result = inverse_laplace_result.rewrite(sp.sin).rewrite(sp.cos)
            inverse_laplace_result = sp.simplify(inverse_laplace_result)
            inverse_laplace_result = inverse_laplace_result.subs(sp.Heaviside(t), 1)
            inverse_laplace_result = inverse_laplace_result.subs(sp.Heaviside(-t), 0)

            return (f"\nInput Function: {f_s}"
                    f"\nPartial Fraction Decomposition: {f_s_decomposed}"
                    f"\nInverse Laplace Transform: {inverse_laplace_result}")
        except Exception as e:
            return f"Error: {e}"

# Main Program
def main():
    print("\nChoose the operation you want to perform:")
    print("- (1) Laplace Transform")
    print("- (2) Inverse Laplace Transform")
    print("Enter 'exit' to quit.\n")

    while True:
        choice = input("Choose operation (1/2 or 'exit'): ").strip().lower()

        if choice == "exit":
            print("Exiting the program. Goodbye!")
            break

        if choice == "1":
            function = input("Enter the function (in terms of 't', e.g., ln(t), t**2): ").strip()
            print(LaplaceTransform.compute(function))
        elif choice == "2":
            function = input("Enter the function (in terms of 's', e.g., 1/(s+2)): ").strip()
            print(InverseLaplaceTransform.compute(function))
        else:
            print("Invalid choice! Please select a valid option.")

        retry = input("\nDo you want to solve another equation? (yes/no): ").strip().lower()
        if retry != "yes":
            print("Exiting the program. Goodbye!")
            break

# Run the program
if __name__ == "__main__":
    main()
```
